1.What is NumPy, and why is it widely used in Python?


 = NumPy is a Python library for fast numerical operations using arrays. It’s widely used for its speed, efficiency, and array support.

2.How does broadcasting work in NumPy?


 = Broadcasting lets arrays of different shapes operate together by expanding the smaller array to match the larger one.

3.What is a Pandas DataFrame?


 = A 2D labeled data structure, like a table, used for storing and analyzing data.

4.Explain the use of the groupby() method in Pandas?

 = groupby() splits data into groups based on keys and applies functions like sum(), mean(), etc.

5.Why is Seaborn preferred for statistical visualizations?

 = It provides clean, easy-to-read plots and integrates well with Pandas for statistical analysis.

6.What are the differences between NumPy arrays and Python lists?

 = NumPy arrays are faster, use less memory, and support math operations; lists are slower and more flexible.

7.What is a heatmap, and when should it be used?

 = A heatmap is a color-coded grid used to visualize data patterns, like correlation matrices.

8.What does the term “vectorized operation” mean in NumPy?

 = It refers to applying operations to entire arrays without using loops, making them faster.

9.How does Matplotlib differ from Plotly?

 = Matplotlib is static and great for reports; Plotly is interactive and better for dashboards.

 = It plots pairwise relationships between numeric columns to explore correlations.

 = Gives quick summary stats (mean, std, min, max, etc.) for numeric columns.

 = Missing data can distort results; handling it ensures cleaner, more accurate analysis.

14.What are the benefits of using Plotly for data visualization?

 = Plotly provides interactive, zoomable, web-friendly charts ideal for presentations and dashboards.

```python
#1.How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print(row_sums)  # Output: [ 6 15 ]
```

```python
#2.Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

df = pd.DataFrame({'scores': [70, 85, 90, 60]})
mean_value = df['scores'].mean()
print(mean_value)  # Output: 76.25
```

```python
#3.Create a scatter plot using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [5, 6, 7, 8]
})

corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#5.Generate a bar plot using Plotly

import plotly.express as px

data = {'Fruits': ['Apples', 'Bananas', 'Cherries'], 'Quantity': [10, 15, 7]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Fruits', y='Quantity', title='Fruit Count')
fig.show()
```

```python
#6.Create a DataFrame and add a new column based on an existing column

import pandas as pd

df = pd.DataFrame({'price': [100, 200, 300]})
df['price_with_tax'] = df['price'] * 1.1
print(df)
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)  # Output: [ 4 10 18 ]
```

```python
#8.Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
```

```python
#9.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

df = pd.DataFrame({'value': [10, 25, 15, 30]})
filtered = df[df['value'] > 20]
print(filtered)
```

```python
#10. Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import matplotlib.pyplot as plt

data = [5, 10, 15, 10, 5, 5, 20, 25, 30]
sns.histplot(data, bins=5, kde=True)
plt.title("Histogram")
plt.show()
```

```python
#11.Perform matrix multiplication using NumPy

import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```

```python
#13.Create a 3D scatter plot using Plotly

import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
